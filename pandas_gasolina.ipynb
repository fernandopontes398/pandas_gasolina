# 4.As colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), para converter ambas colunas para Timestamp / Datetime.

```python
df_combinado['DATA INICIAL'] = pd.to_datetime(df_combinado['DATA INICIAL'], errors='coerce')
df_combinado['DATA FINAL'] = pd.to_datetime(df_combinado['DATA FINAL'], errors='coerce')

df_combinado.info()
```

#  5. Criando uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência.

```python
df_combinado["ANO-MES"] = df_combinado["DATA FINAL"].apply(lambda x: "{}".format(x.year)) + df_combinado["DATA FINAL"].apply(lambda x: "/{:02d}".format(x.month))
```

#### 7. Filtrando o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Gravando em uma nova variável.

#### 8. Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
preco_medio_agosto_2008 = gasolina_comum_df[gasolina_comum_df['ANO-MES'] == '2008/08']['PREÇO MÉDIO REVENDA'].mean()
print(f"O preço médio de revenda da gasolina comum em agosto de 2008 foi de: R$ {preco_medio_agosto_2008:.3f}")
```

# Esse valor está acima ou abaixo da média anual de 2008?

```python
preco_medio_anual_2008 = gasolina_comum_df[gasolina_comum_df['DATA INICIAL'].dt.year == 2008]['PREÇO MÉDIO REVENDA'].mean()
preco_medio_agosto_2008 = gasolina_comum_df[gasolina_comum_df['ANO-MES'] == '2008/08']['PREÇO MÉDIO REVENDA'].mean()
diferenca = (preco_medio_agosto_2008 - preco_medio_anual_2008) / preco_medio_anual_2008 * 100
print(f"O preço médio de revenda da gasolina comum no ano de 2008 foi de: R$ {preco_medio_anual_2008:.3f}")
print(
    f"Em agosto de 2008, o preço médio da gasolina foi "
    f"{abs(diferenca):.3f}% {'abaixo' if diferenca < 0 else 'acima'} "
    f"da média anual."
)
```

O preço médio de revenda da gasolina em agosto de 2008 apresentou valor muito próximo à média anual do mesmo ano, com variação inferior a 1%. Isso indica que agosto não se destacou como um mês atípico, sugerindo estabilidade dos preços nesse período.

#### 12. Uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro

```python
dados_rio_de_janeiro = gasolina_comum_df[gasolina_comum_df['ESTADO'] == 'RIO DE JANEIRO'].copy()

preco_medio_anual_rj = dados_rio_de_janeiro.groupby(dados_rio_de_janeiro['DATA INICIAL'].dt.year)['PREÇO MÉDIO REVENDA'].mean()

variacao_percentual_anual = preco_medio_anual_rj.pct_change() * 100

resultado_rj = pd.DataFrame({'Preço Médio Anual (R$)': preco_medio_anual_rj,'Variação Percentual Anual (%)': variacao_percentual_anual})

print("Variação percentual ano a ano do preço médio da gasolina no Rio de Janeiro:")
display(resultado_rj)
```

# O RJ apresenta maior volatilidade que a média nacional?

```python
dados_nacional = gasolina_comum_df

preco_medio_anual = dados_nacional.groupby(dados_nacional['DATA INICIAL'].dt.year)['PREÇO MÉDIO REVENDA'].mean()

variacao_percentual = preco_medio_anual.pct_change() * 100

resultado = pd.DataFrame({'Preço Médio Anual (R$)': preco_medio_anual,'Variação Percentual Anual (%)': variacao_percentual})

print("Variação percentual ano a ano do preço médio da gasolina nacional:")
display(resultado)

std_rj = variacao_percentual_anual.std()
std_nacional = variacao_percentual.std()

print(f"\nDesvio padrão da variação percentual anual no Rio de Janeiro: {std_rj:.2f}%")
print(f"Desvio padrão da variação percentual anual nacional: {std_nacional:.2f}%")

if std_rj > std_nacional:
    print("O Rio de Janeiro apresenta maior volatilidade que a média nacional.")
elif std_rj < std_nacional:
    print("O Rio de Janeiro apresenta menor volatilidade que a média nacional.")
else:
    print("O Rio de Janeiro apresenta volatilidade semelhante à média nacional.")
```

#### Criando uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros.Também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
# Agrupar por 'ANO-MES' para encontrar os preços mínimo e máximo e os estados correspondentes
min_max_prices = gasolina_comum_df.groupby('ANO-MES').agg(
    min_preco=('PREÇO MÉDIO REVENDA', 'min'),
    max_preco=('PREÇO MÉDIO REVENDA', 'max')
).reset_index()

# Função para obter o estado com o preço mínimo/máximo
def get_state_for_price(df, year_month, price_col, price_value):
    temp_df = df[(df['ANO-MES'] == year_month) & (df[price_col] == price_value)]
    if not temp_df.empty:
        return temp_df['ESTADO'].iloc[0] # Retorna o primeiro estado encontrado
    return None

# Adicionar os estados correspondentes aos preços mínimo e máximo
min_max_prices['estado_menor_preco'] = min_max_prices.apply(lambda row: get_state_for_price(gasolina_comum_df, row['ANO-MES'], 'PREÇO MÉDIO REVENDA', row['min_preco']), axis=1)
min_max_prices['estado_maior_preco'] = min_max_prices.apply(lambda row: get_state_for_price(gasolina_comum_df, row['ANO-MES'], 'PREÇO MÉDIO REVENDA', row['max_preco']), axis=1)

# Calcular a diferença absoluta e percentual
min_max_prices['diferenca_absoluta'] = min_max_prices['max_preco'] - min_max_prices['min_preco']
min_max_prices['diferenca_percentual'] = (min_max_prices['diferenca_absoluta'] / min_max_prices['min_preco']) * 100

# Exibir a tabela resultante
print("Tabela de Variação de Preços (Mínimo vs. Máximo) por Ano-Mês:")
display(min_max_prices.head())
```
